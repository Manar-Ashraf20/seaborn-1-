# relplot function() & ScatterPlot

relplot function() =>  visualize the relationship between two quantitative variables using either scatter plots or line plots.

#####  scatterplot() using  relplot()

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter")
```

##### Subplots in columns  => using 'col' parameter

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
           col="smoker")
```

##### subplots in rows ===> using 'row' parameter 

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
           row="smoker")
```

##### Subplots in rows and columns

فى حالة انى هضيف متغيرين اضافين 

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
           col="smoker",
           row="time")
```

##### Wrapping columns (تحديد عدد ال subplots فى كل صف) 
فى حالة ان القيم كتيرة فى الصف الواحد

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
           col="day")
```

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
           col="day",
           col_wrap=2)
```

###### Ordering columns

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
           col="day",
           col_wrap=2,
           col_order=["Thur","Fri","Sat","Sun"])
```

# Boxplot

 compare the distribution of quantitative variable across different groups of a categorical variable 

 colored box =>  represents the 25th to 75th percentile
 line in the middle of the box => represents the median
 whiskers => give the sence of spread of the distribution 
 floating point => represent outliers 

### Create a box plot

```python
sns.catplot(x="time",
            y="total_bill",
            data=tips,
            kind="box")
plt.show()
```

### Changing order of categories using 'order' parameter 

```python
sns.catplot(x="time",y="total_bill",data=tips,kind="box",order=["Dinner","Lunch"])
```

### Omit outliers using 'showfliers' parameter 

```python
g = sns.catplot(
    x="time", 
    y="total_bill", 
    data=tips, 
    kind="box", 
    showfliers=False 
)
plt.show()
```

### IQR & Whiskers

 IQR => is the 25th to 75th percentile of a distribution 
 whiskers extend to 1.5* the IQR 

 changing it with whis parameter 
 whis=2.0
 or 
 whis=[5,95] او اى نسبة انتى عاوزها 

```python
sns.catplot(x="time",y="total_bill",data=tips,kind="box",whis=[0,100])
plt.show()
```

```python
sns.catplot(x="time",y="total_bill",data=tips,kind="box",whis=[5,95])
plt.show()
```

```python
sns.catplot(x="time",y="total_bill",data=tips,kind="box",whis=2)
plt.show()
```

### Changing the figure style using '   .set_style()' function 

5 styles which change the background & axes of the plot 

1) white => provides clean axes with a solid white background. 
2) whitegrid => will add a gray grid in the background. 
3) ticks =>  similar to "white", but adds small tick marks to the x- and y-axes.
4) dark =>  provides a gray background
5) darkgrid => provides a gray background with a white grid

```python
sns.set_style('darkgrid')
sns.catplot(x="day" , y="total_bill" , data = tips ,kind="point")
plt.show()
```

### Changing the palette => using '    hue & legend=False & palette ' parameters 

1) present palettes 
2) create a custom palette 

##### present palettes

1) Diverging palettes ( RdBu & RdBu_r     &     PRGn & PRGn_r)
2) sequential palettes => single color moving from light to dark 
                       => جيدة للتأكيد على متغير على نطاق مستمر

##### Add a title to a FacetGrid object

1) assign the plot to the variable "g".
2) set the title using "g.fig.suptitle()"

```python
g=sns.catplot(x="time",y="total_bill",data=tips,kind="box")
g.fig.suptitle("new title")
plt.show()
```

##### Adjusting height of title in FacetGrid => using the "y" parameter.


```python
g=sns.catplot(x="time",y="total_bill",data=tips,kind="box")
g.fig.suptitle("new title",y=0.9)
plt.show()
```

##### Adding a title to AxesSubplot & ( Adjusting height of title using "y" parameter)

1) assign the plot to the variable "g"
2) set the title using "g.set_title()"

```python
g=sns.boxplot(x="time",y="total_bill",data=tips)
g.set_title("new title",y=0.9)
plt.show()
```

##### Titles for subplots 

1) using "g.fig.suptitle()"
              => to add a title to the figure as a whole      (FacetGrid object)
2) g.set_titles()" 
              =>  to set the titles for each AxesSubplot.
    (If you want to use the variable name in the title,
        you can use "col_name" in {} to reference the column value) 

```python
g=sns.catplot(x="time",y="total_bill",data=tips,kind="box",col="sex")
g.fig.suptitle("new title",y=1.03)
g.set_titles(" this is {col_name}")
```
